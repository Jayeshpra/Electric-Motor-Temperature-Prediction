```python
import pandas as pd
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
Dataframe = pd.read_csv("D:\\Jayesh\\Git\\GitHub repos\\Electric-Motor-Temperature-Prediction\\Dataset.csv") # write the path to your dataset here
Dataframe
```

```python
X = Dataframe[['i_q', 'i_d', 'u_q', 'u_d', 'ambient', 'motor_speed', 'coolant']]
y = Dataframe['pm']
display(X)
display(y)
```

```python
X.isnull().sum()
```

```python
mm_scaler_x = MinMaxScaler()
X = mm_scaler_x.fit_transform(X)
X = pd.DataFrame(X, columns=['i_q', 'i_d', 'u_q', 'u_d', 'ambient', 'motor_speed', 'coolant'])
display(X)
```

```python
joblib.dump(mm_scaler_x, 'mm_scaler_x.save')
```

```python
mm_scaler_y = MinMaxScaler()
y = mm_scaler_y.fit_transform(pd.DataFrame(y))
y = pd.DataFrame(y, columns=['pm'])
display(y)
```

```python
mm_scaler_y.inverse_transform(y)
```

```python
joblib.dump(mm_scaler_y, 'mm_scaler_y.save')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
decision_tree_model = DecisionTreeRegressor()
random_forest_model = RandomForestRegressor(n_estimators=50)
svr_model = SVR()
```

#### You can ignore random_forest_model and support_vactor_model as they take too long to be trained. we can get 98% accuracy using decision tree model which is easy to train.

```python
linear_model.fit(X_train, y_train)
```

```python
decision_tree_model.fit(X_train, y_train)
```

```python

random_forest_model.fit(X_train, y_train.values.ravel()) 
```

```python
print(f"r2 score for linear model: {r2_score(y_test, linear_model.predict(X_test))}")
print(f"r2 score for decision tree model: {r2_score(y_test, decision_tree_model.predict(X_test))}")
print(f"r2 score for random forest model: {r2_score(y_test, random_forest_model.predict(X_test))}")
```

```python
print(f"mean squared error for linear model: {mean_squared_error(y_test, linear_model.predict(X_test))}")
print(f"mean squared error for decision tree model: {mean_squared_error(y_test, decision_tree_model.predict(X_test))}")
print(f"mean squared error for random tree model: {mean_squared_error(y_test, random_forest_model.predict(X_test))}")
```

```python
print(f"mean absolute error for linear model: {mean_absolute_error(y_test, linear_model.predict(X_test))}")
print(f"mean absolute error for linear model: {mean_absolute_error(y_test, decision_tree_model.predict(X_test))}")
print(f"mean absolute error for linear model: {mean_absolute_error(y_test, random_forest_model.predict(X_test))}")
```
